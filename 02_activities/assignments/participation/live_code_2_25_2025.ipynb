```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0, 100,50)
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y)
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.bar(x,y)
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.hist(y)
```

```python
font1 = {'family':'sans-serif','color':'blue','size':20}  # Font style for title
font2 = {'family':'monospace','color':'green','size':14}  # Font style for labels
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)

ax.set_title('Total growth over time', fontdict=font1, loc='left') #adding title with custom font
ax.set_ylabel('Total growth', fontdict=font2)  # adding y-axis label with custom font
ax.set_xlabel('Years since start', fontdict=font2) # adding x-axis label with custom font
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter( x,
y,
marker='*' # Star-shaped markers for visibility
, 
color = "indigo") # Set color to ensure consistency
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,
y,
marker='*'
,
color = 'indigo'
,
linestyle = '--'
, 
linewidth = 2) 
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,
y,
marker='.' # Use dot markers for better visibility
,
color = 'blue' # Line color
,
linestyle = '--' # Dashed line style
, 
markersize= 12, # Size of the markers
markeredgecolor = '#fa9359',  # Edge color of the markers (orange shade)
markerfacecolor = 'pink', # Fill color of the markers
linewidth = 2)  # Width of the line

# Customize the grid for better readability
ax.grid(color = "red"
, linewidth = '1', 
linestyle = '--')
```
